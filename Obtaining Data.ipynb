```python
def scrape_tweets(company_name):
    tweets_list1b = []

    # Using TwitterSearchScraper to scrape data and append tweets to list
    for i in range(delta.days + 1):
        string = str(company_name)+ ' since:' + str(d1 + timedelta(days=i)) + ' ' + 'until:'+ str(d1 + timedelta(days=i+1))
        print(string)
        for j,tweet in enumerate(sntwitter.TwitterSearchScraper(string).get_items()):
            if j>10:
                break
            tweets_list1b.append([tweet.date, tweet.id, tweet.rawContent, tweet.user.username])
    
    # Creating a dataframe from the tweets list above
    tweets_df1b = pd.DataFrame(tweets_list1b, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
    
    output_string = str(company_name) + '_stocktweets.csv'
    
    tweets_df1b.to_csv(output_string, index = False)
```

```python
scrape_tweets('Proctor & Gamble')
```
